```python
import pandas as pd
# Load Data
df = pd.read_csv("../data/winemag-data-130k-v2.csv")
```

```python
# Task: Who are the most common wine reviewers.
# Create a series of whose index is the 'taster_twitter_handle'
# Whose values count how many reviews each person wrote
# Method 1
taster_review_count = df.groupby('taster_twitter_handle').size()
# Method 2: All columns
# taster_review_count = df.groupby('taster_twitter_handle').count()
# Method 3: One column
#taster_review_count = df.groupby('taster_twitter_handle')['taster_twitter_handle'].count()
print(taster_review_count)
```

```python
# Task: What is the best wine I can buy for a given amount of money?
# Create a Series of whose index is the wines price and whose values are the max number of points
# a wine costing that much was given in a review
# Sort the series by price, ascending ($4 at the top and #3300 at the bottom)
best_wine_by_price = df.groupby('price').points.max().sort_index()
print(best_wine_by_price)
```

```python
# Task: What are the minimum and maximum prices for each variety of wine?
# Create a Series of whose index is the 'variety' and whose values are the min and max val
min_max_price_by_variety = df.groupby('variety').price.agg(['min', 'max'])
min_max_price_by_variety
```

```python
# Task: What are the most expensive wine varieties?
# Create a variable 'sorted_varieties' containg a copy of the 
# DataFrame from the PREVIOUS task
sort_varieties = min_max_price_by_variety.sort_values(by=['max', 'min'], ascending=False)
sort_varieties
```

```python
# Task: Create a series whose index is 'reviewers' and whose values are
# the average review score given by the reviewer. Sort from higher to lower
# Hint: you need 'taster_name' and 'points'
avg_score_by_reviewer = df.groupby('taster_name').points.mean().sort_values(ascending=False)
print(avg_score_by_reviewer)
```
