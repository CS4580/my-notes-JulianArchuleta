```python
import pandas as pd
import matplotlib.pyplot as plt
# Display the plots inside the notebook
%matplotlib inline
```

```python
# Load the data
homes_df = pd.read_csv('../data/homes.csv')
homes_df.head()
```

### Scatter Plots
Useful for visualizing relationships between two data points

```python
# Plot the data
homes_df.plot()
```

```python
# Specify the x and y axis labels
homes_df.plot(x='List', y='Rooms')
```

```python
# Add markers
homes_df.plot(x='List', y='Rooms', marker='o', markerfacecolor='r')
```

### Bar Plot
Useful for when you are showing or comparing segments of information or  categorical data

```python
# Load data
cites_df = pd.read_csv('../data/cities.csv')
cites_df.head()
```

```python
# Filter recors from Year = 2019
cites_2019_df = cites_df[cites_df['Year'] == 2019]
cites_2019_df.plot()
```

```python
# Now plot a bar chart of the number of cities in each region
cites_2019_df.plot(kind='bar', x='City', y='Population', title='Population of Cities in 2019', figsize=(8,4), color='teal')
```

```python
# You may want to plot data horizontally
cites_2019_df.plot(kind='barh', x='City', y='Population', title='Population of Cities in 2019', figsize=(8,4), color='orange', legend=False)
```

### Scatter Plots
When you are interested in the relationship between two variables, scatter plots are probably the best choice

```python
# Load the data
houses_df = pd.read_csv('../data/homes.csv')
houses_df.head()
```

```python
# Plot the scatter plot
houses_df.plot(kind='scatter', x='List', y='Rooms')
```

```python
# Filter the data directly
houses_df[houses_df['Age'] < 50].plot(kind='scatter', x='Age', y='List')
```

```python
# If you wat to eliminate outliers, you can filter the data before plotting
houses_plot_df = houses_df[houses_df['Age'] < 50]
houses_plot_df = houses_plot_df[houses_plot_df['List'] < 350]
# Plot the data
houses_plot_df.plot(kind='scatter', x='Age', y='List')
```

### Area Plot
Useful when you want to understand how values develop and are composed over time

```python
# Load the data set
enrollments_df = pd.read_csv('../data/enrollment.csv')
enrollments_df.head()
```

```python
# Create an area plot
enrollments_df.plot(kind='area', x='Year', figsize=(8,4))
```

```python
# Data is not good because it is not organized by year
# We want to see the total, and how it changes over time
# Sort the data by year, then plot it using the pivot method in pandas
pivot_df = enrollments_df.pivot(index='Year', columns='Class', values='Enrollment')
pivot_df.head()
```

```python
# Plot it
pivot_df.plot(kind='area', figsize=(8,4))
```

```python
# The data is stacked by deafult, if you want to see it side by side
# You can set this option to False, stacked= False
pivot_df.plot(kind='area', figsize=(8,4), stacked=False)
```
