## Introduction to Plotting

### Python Packages for plotting
- `matplotlib` : All plots in oabdas are matplotlib objects
- `seaborn` : Focused on statistical visualization
- `bokeh`: To create interactive visualizations for the internet 
- `plotly`: Create interactive and embeddable visualizations 

### Matplotlib
Main submodule is `pyplot` which will give you different functions to work with different plot elements. 

### Creating a plot
- Import the necessary packages
- Import or create the data
- Call the apporiate plotting functions

```python
import pandas as pd
import matplotlib.pyplot as plt
# Display the plots inside the notebook
%matplotlib inline
```

```python
# Dummy data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
print(x)
print(y)
```

```python
# Default is line plot
plt.plot(x, y)
```
