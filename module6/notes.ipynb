## Task: Understand Museum Attendance

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# load data 
file_path = '../data/museum_visitors.csv'
museum_df = pd.read_csv(file_path, index_col="Date", parse_dates=True)

museum_df.head()
```

The Firehouse museum claims they ran an event in 2014 that brought a big number of vistors, and that they should an extra budget to run a similar event this year.

Task: Show the museum board how the event compard to regular traffic at each museum.
- Create a `lineplot chart` that shows how the number of visitors change on each museum over time
- Provide a title for the plot

```python
# plot the data
plt.figure(figsize=(16,6))
plt.title("Daily Visitors to Museums")

sns.lineplot(data=museum_df)
```

### Assess Seasonality
When meeting with the employees of the `Avila Adobe`, you hear that one major pain point is that the number of the vistors varies greatly with the season.

Part A:
- Create a `line chart` that shows how the number of visitors to `Avila Adobe` has evolved over time
- Replace `inf` with `nan`, and then fill them up in `0`

```python
import numpy as np
# Replace infinitie values with NaN
museum_df = museum_df.replace([np.inf, -np.inf], np.nan)
# Now we can use the fillna() method to replace NaN values with zeros
museum_df = museum_df.fillna(0)
```

```python
# plot the data
plt.figure(figsize=(16,6))
plt.title("Monthly Visitors to Avila Adobe")

sns.lineplot(data=museum_df['Avila Adobe'], label="Avila Adobe")
```

## Bar Plots
Dataset: US Department of Transportation that tracks flight delays

```python
# load dataset
file_path = '../data/flight_delays.csv'
flight_delays_df = pd.read_csv(file_path, index_col="Month")

flight_delays_df.head()
```

```python
# use a barplot
plt.figure(figsize=(14,7))	
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

sns.barplot(x=flight_delays_df.index, y=flight_delays_df['NK'])
```

## Heat Maps

```python
# plot heatmap for all data
plt.figure(figsize=(16,6))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=flight_delays_df, annot=True)
plt.xlabel("Airline")
```
